```
import os
import nltk
from nltk.tokenize import word_tokenize
from tqdm import tqdm
import numpy as np
import gensim
import string
import pandas as pd

sentences = []
for d in data:
  sentences.append(d['commit']['message'])

print(sentences)
```

```
from gensim.models.doc2vec import Doc2Vec
model = Doc2Vec.load('doc2vec.bin')
processed_sentences = []
for sentence in sentences:
  processed_sentences.append(word_tokenize(sentence.lower()))
  
```

```
sentence_vectors = []
for sentence in processed_sentences: 
  sentence_vectors.append(model.infer_vector(sentence))
```

```
bug = 'broken dropdowns is not working and unable to start'
preprocessed_bug = word_tokenize(bug.lower())
bug_vector = model.infer_vector(preprocessed_bug)
```

```
indexes = []
i = 0
while i<len(sentence_vectors):
  indexes.append((np.dot(bug_vector,sentence_vectors[i]),i))
  i+=1

indexes = sorted(indexes)
print(indexes)
i = len(indexes)-1
count = 1
while count<=5:
  print(count, sentences[indexes[i][1]])
  print()
  i-=1
  count+=1
```
